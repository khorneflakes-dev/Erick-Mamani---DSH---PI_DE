```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
path = '/content/drive/MyDrive/test/Dataset Yelp/'
```

```python
df_business = pd.read_json(f'{path}business.json', lines=True)
```

```python
def dict_to_columns(df_i, column):
  df_i[column] = df_i[column].fillna('{}')
  df_i.reset_index(inplace=True)
  df_o = df_i.join(pd.json_normalize(df_i.pop(column)))
  if 'level_0' in df_o.columns:
    df_o.drop(['level_0'], axis=1, inplace=True)

  return df_o
```

```python
df_business_cl = dict_to_columns(df_business, 'attributes')
df_business_cl = dict_to_columns(df_business_cl, 'hours')
df_business_cl = dict_to_columns(df_business_cl, 'BusinessParking')
```

```python
for i in df_business_cl:
  di = {'True':1, True: 1, 'False': 0, False: 0, np.nan: 0, 'None': 0}
  df_business_cl[i].replace(di, inplace=True)
```

```python

df_demo = df_business_cl[df_business_cl.columns[13:-1]]

for i in df_demo.columns:
  print(df_demo[i].value_counts())
```
